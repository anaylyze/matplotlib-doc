## Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import data

```python
cs_df = pd.read_csv('https://raw.githubusercontent.com/derekbanas/seaborn/refs/heads/master/ComputerSales.csv')
cs_df.head()
```

### Built-In Seaborn Datasets

```python
print(sns.get_dataset_names())
```

```python
crash_df = sns.load_dataset('car_crashes')
crash_df
```

### DISTRIBUTION PLOTS

#### Distribution Plot

```python
sns.distplot(crash_df['not_distracted'],kde=False, bins=25)
# ked-> kernel density estimation
```

### Joint Plot

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hex')
```

### KDE Plot

```python
sns.kdeplot(crash_df['alcohol'])
```

### Pair Plots

```python
tips_df = sns.load_dataset('tips')

sns.pairplot(tips_df, hue='sex', palette='Blues')
```

#### Rug Plot

```python
sns.rugplot(tips_df['tip'])
```

### Styling

```python
sns.set_style('ticks') #used to set style of the jointplot

plt.figure(figsize=(8,4)) #used to set size

sns.set_context('paper',font_scale=2) #to change the labels and lines and other diff elements

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg') #to make a jointplot

sns.despine(left=True, right=True) # to disable any label or function of the plot
```

# Categorical Plots

## Bar Plots

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.cov)
# estimator can be used to change on what basis the plot is
# default it is based on mean
```

 ## Count Plot

```python
sns.countplot(x='sex',data=tips_df)
```

## Box Plot

```python
# it is used to compare diff kind of variables
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')

```

## Violin Plot

```python
sns.violinplot(x='day',y='total_bill',data=tips_df, hue='sex', split=True)
```

## Strip Plot

```python
plt.figure(figsize=(8,5))

sns.stripplot(x='day',y='total_bill', data=tips_df, jitter=True, hue='sex', dodge=True)
#jitter is used to scatter the points so they don't get stacked over eachother
# hue is used to separate the data for each day
# Dodge is used to separate the data further after the sex separation in the o/p
```

## Swarm Plot

```python
sns.violinplot(x='day', y='total_bill', data=tips_df)
sns.swarmplot(x='day', y='total_bill', data=tips_df, color='white')
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips_df)
```

## Palettes

```python
plt.figure(figsize=(8,5))
sns.set_style('white')
sns.set_context('talk')
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex', palette='afmhot')

# check matplotlib documentation for colormaps

plt.legend(loc='upper left', bbox_to_anchor=(1,1))
```
