```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#Resize your graph
plt.figure(figsize=(10,5), dpi=100) #dpi is pixels per inch

plt.plot(x,y, label='2x', color='black', linewidth=2,linestyle='--', marker='*',markersize=10, markeredgecolor='yellow') #we can also pass hex code


# use of shorthand notation
# fmt='[color][marker][line]'
# plt.plot(x,y,'b^--', label='2x')

# line number two
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,'y-', label='x^2') #making some a fancy line of two different types in a single one
plt.plot(x2[5:],x2[5:]**2,'y^--')

plt.title('Our Frist Graph', fontdict={'fontname':'Comic Sans MS', 'fontsize':'25'})
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.xticks([0,1,2,3,4,4.5]) #it is used to set the ticklemarks(scale of your graph) on the axes
plt.yticks([0,2,4,6,8,10])

plt.legend() #to add a legend

plt.savefig('mygraph.png',dpi=300) #to save the figure, dpi 300 is good when saving so graph has high resolution

plt.show() #to get rid of  the line [<matplotlib.lines.Line2D at 0x1b09275cce0>]
```

### Bar Chart

```python
labels = ['A','B','C']
values = [1,4,2]

bars = plt.bar(labels, values)

patterns = ['/','+','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))


# bars[0].set_hatch('/')
# bars[1].set_hatch('+')
# bars[2].set_hatch('*')

plt.figure(figsize=(6,4))



plt.show()
```
