# Real World Examples

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Line Graph

```python
gas = pd.read_csv('gas_prices.csv')

# making the graph bigger
plt.figure(figsize=(8,5))

# adding a title
plt.title('Gas Prices Overtime (in $)', fontdict={'fontname':'Comic Sans MS','fontweight':'bold','fontsize':'17'})

plt.plot(gas.Year, gas.USA,'g.-', label='United States of America')
plt.plot(gas.Year, gas.Japan,'y.-',label='Japan')
plt.plot(gas.Year, gas.Germany, 'b.-', label='Germany' )

# plotting for multiple word "WORD"
plt.plot(gas['Year'], gas['South Korea'],'r.-', label='South Korea')

# Another way to plot is
# for country in gas:
#     if country != 'Year':
#         plt.plot(gas.Year,gas[country], marker='.', label=country)

# Changing the tick marks for clean and accurate axes
# plt.xticks(gas.Year[::3]) 

# If you want to add values to axes as originally it was terminating on 2008
plt.xticks(gas.Year[::3].tolist()+[2011])

# labelling the axes
plt.xlabel('Year',fontdict={'fontname':'Comic Sans MS'})
plt.ylabel('US Dollar($)', fontdict={'fontname':'Comic Sans MS'})

# Adding legend so the user can understand the graph
plt.legend(loc='upper left', bbox_to_anchor=(1,1))

# to save the graph
plt.savefig('Gas_Price_Overtime.png', dpi=300)

plt.show()
```

# Fifa Data

```python
fifa = pd.read_csv('fifa_data.csv')
```

```python
fifa.sort_values(by='Overall',ascending=False)
```

### Histograms

```python
# to set bins on the histogram
plt.figure(figsize=(8,5))

bins= [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins, color='#51414F')

plt.xticks(bins)

plt.title('Distribution of players on their Rating in FIFA 2018', fontdict={'fontname':'Comic Sans MS','fontsize':'30','fontweight':'bold'})
plt.ylabel('No. of players',fontdict={'fontname':'Comic Sans MS','fontsize':'10','fontweight':'bold'})
plt.xlabel('Rating of players', fontdict={'fontname':'Comic Sans MS','fontsize':'10','fontweight':'bold'})




plt.show()
```

### Pie Chart

```python
# making a pie chart for right footed v/s left footed players


left = fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]

labels = ['Left','Right']
colors = ['#d9d9f3','#d62020']

plt.pie([left, right], labels=labels,colors = colors, autopct='%.2f %%')
plt.title('Strong foot of players')
plt.show()
```

### Pie Chart

#### Pie chart on weight of footballers

```python
#extracting weight from string
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 170)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight >= 200].count()[0]
```

```python
weights = [light, light_medium, medium, medium_heavy, heavy]

labels = ['Under 125','125-150','150-175','175-200',' Over 200']

plt.style.use('ggplot') #can use this to change colours

# what explode does is that it splits the pie apart so it not touching
explode =  (.4,.1,0,0,.4)

plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)
# pctdistance sets the distance of percentage from the centre

plt.title('Weight Distrubution of FIFA Players (in lbs)')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Example: if you are reading the DataFrame
# fifa = pd.read_csv('fifa_data.csv')

# Make sure `fifa` DataFrame contains the necessary columns and data
barcelona = fifa.loc[fifa['Club'] == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa['Club'] == 'Real Madrid']['Overall']
ajax = fifa.loc[fifa['Club'] == 'Ajax']['Overall']
alh = fifa.loc[fifa['Club'] == 'Al Hilal']['Overall']

labels = ['FC Barcelona','Real Madrid CF','Ajax','Al-Hilal']

# Plot the boxplot
plt.style.use('default')
plt.figure(figsize=(5,8))

# Create the boxplot
boxes = plt.boxplot([barcelona, madrid, ajax, alh], labels=labels, patch_artist=True, medianprops={'linewidth':2})

# Customize the boxplot
for box in boxes['boxes']:
    box.set(color='#4286f4', linewidth=2)  # Set edge color
    box.set(facecolor='#F4C2C2')  # Set fill color

# Add title and labels
plt.title('Football Team Comparison based on FIFA Rating')
plt.ylabel('FIFA OVR')

# Show plot
plt.show()
```
